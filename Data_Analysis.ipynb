```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
df
```

```python
df['Q1'] = df['Q1'].replace('PhD', 'phd')
df['Q1'] = df['Q1'].replace('Phd', 'phd')
Course_counts = df['Q1'].value_counts()
print(Course_counts)
plt.figure(figsize=(8, 6))
Course_counts.plot(kind='bar')
plt.title('Number of Entries by Course')
plt.xlabel('Course')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()

```

```python
df['Q2'] = df['Q2'].replace('17-22 ', '17-22')
df['Q2'] = df['Q2'].replace('23-28 ', '23-28')
Age_counts = df['Q2'].value_counts()

print(Age_counts)

plt.figure(figsize=(8, 6))
Age_counts.plot(kind='bar')
plt.title('Number of Entries by Age')
plt.xlabel('Age')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

```python
df['Q3'] = df['Q3'].replace('First Year ', 'First Year')
Year_counts = df['Q3'].value_counts()

print(Year_counts)

plt.figure(figsize=(8, 6))
Year_counts.plot(kind='bar')
plt.title('Number of Entries by Year')
plt.xlabel('Year')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

```python
df['Q4'] = df['Q4'].replace('CIVIL ', 'CIVIL')
df['Q4'] = df['Q4'].replace('CHEM ', 'CHEM')
Branch_counts = df['Q4'].value_counts()

print(Branch_counts)
print(Branch_counts.mean())
plt.figure(figsize=(8, 6))
Branch_counts.plot(kind='bar')
plt.title('Number of Entries by Branch')
plt.xlabel('Branch')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

```python
Gender_counts = df['Q5'].value_counts()
print(Gender_counts)
plt.figure(figsize=(8, 6))
Gender_counts.plot(kind='bar')
plt.title('Number of Entries by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```
